```
def getNextMovie(s):
    for i in range(0,1000):
      if (movieData)[i][10] == s[10] or (movieData)[i][9] == s[1] or (movieData)[i][9] == s[9]:
        if movieData[i][11] != s[11]:
          return movieData[i]
        else:
          return random.choice(movieData)
      else: 
        return random.choice(movieData)
```
